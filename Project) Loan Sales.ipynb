## Project) Loan Sales 

#### In this project you are tasked wth analyzing this dataset to gather insights  for the Department Manager for Loans Sales.

#### The Department Manager wants insights on the following :
#### - Reason why people applied for loans
#### - The average: Fico_Score, loan amount, gross income, and Housing payment for applicants approved and denied
#### - Employment Status for each applicant approved and denied


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Personal Loans Dataset.csv")
df.head()
```

```python
df['Approved']=df.Approved.apply(lambda x: "Approved" if x == 1 else("Denied" if x==0 else x))
df.drop(columns='Column1',inplace=True)
```

### Reason )
#### Whats the most commmon Reason for each Loan ?
####  Whats the most common Reason for each Applicant that was Approved  ?
####  Whats the most common Reason for each Applicant that was Denied  ?

```python
df.Reason.value_counts()
```

```python
df.loc[df.Approved=='Approved'].groupby(['Reason','Lender'])[['applications']].count()
```

```python
df.loc[df.Approved=='Denied'].groupby(['Reason','Lender'])[['applications']].count()
```

### FICO_score )
#### Whats the average FICO_score for each Lender ?
####  Whats the average FICO_score for each lender that was Approved  ?
####  Whats the average FICO_score for each lender that was Denied  ?

```python
df.groupby('Lender')['FICO_score'].agg(['mean','count'])
```

```python
df.groupby(['Lender','Approved'])['FICO_score'].agg(['mean','count']).round(decimals=1)
```
